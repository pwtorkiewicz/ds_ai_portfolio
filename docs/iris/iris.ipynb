# EDA - 3 krok: Analiza pojedynczych zmiennych wraz z transformacją danych

* 'długość kielicha (sepal length)'
* 'szerokość kielicha (sepal width)'
* 'długość płatka (petal length)'
* 'szerokość płatka (petal width)'
* 'dł/szer_kielicha' - nowy parametr, współczynnik długości do szerokości kielicha
* 'dł/szer_płatka'  - nowy parametr, współczynnik długości do szerokości płatka

## Transformacja danych
- Dodano dwie kolumny:
  - współczynnik długości do szerokości kielicha
  - współczynnik długości do szerokości płatka

```python
df['dł/szer_kielicha']=df['długość kielicha (sepal length)']/df['szerokość kielicha (sepal width)'] # Dodanie nowej kolumny - współczynnika długości do szerokości kielicha
df['dł/szer_płatka']=df['długość płatka (petal length)']/df['szerokość płatka (petal width)']       # Dodanie nowej kolumny - współczynnika długości do szerokości płatka
df.sample(5)
```
